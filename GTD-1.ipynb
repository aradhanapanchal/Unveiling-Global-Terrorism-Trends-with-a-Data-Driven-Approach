```python
#Number of terrorist attacks over the years
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(12, 6))
sns.countplot(x='Year', data=df)
plt.title('Number of Terrorist Attacks Over the Years')
plt.xlabel('Year')
plt.ylabel('Number of Attacks')
plt.xticks(rotation=45)
plt.show()
```

```python
#types of Attack 
plt.figure(figsize=(10, 6))
attack_types = df['Attack Type'].value_counts()
sns.barplot(x=attack_types.index, y=attack_types.values)
plt.title('Number of Terrorist Attacks by Attack Type')
plt.xlabel('Attack Type')
plt.ylabel('Number of Attacks')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
#Weapons usage 
plt.figure(figsize=(10, 6))
weapons_used = df['Weapon Type'].value_counts()
sns.barplot(x=weapons_used.index, y=weapons_used.values)
plt.title('Number of Terrorist Attacks by Weapon Type')
plt.xlabel('Weapon Type')
plt.ylabel('Number of Attacks')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
import geopandas as gpd
import matplotlib.pyplot as plt

#load world map data
world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))

# here, we have grouped data by country and number of terrorist attacks
country_attacks = df['Country'].value_counts().reset_index()
country_attacks.columns = ['name', 'Attacks']
world = world.merge(country_attacks, how='left', on='name')

# Plot map for location of terrorist attacks
fig, ax = plt.subplots(1, 1, figsize=(15, 10))
world.plot(column='Attacks', cmap='OrRd', linewidth=0.8, ax=ax, edgecolor='0.8', legend=True)
ax.set_title('Geographical analysis of Terrorist Attacks')
plt.show()
```

```python
# Group data by terrorist group and count number of attacks
terrorist_groups = df['Group Name'].value_counts().reset_index().head(10)
terrorist_groups.columns = ['Terrorist Group', 'Number of Attacks']

#Plot bar graph for top Terrorist Organizations
plt.figure(figsize=(12, 6))
sns.barplot(x='Number of Attacks', y='Terrorist Group', data=terrorist_groups)
plt.title('Top 10 Active Terrorist Organizations')
plt.xlabel('Number of Attacks')
plt.ylabel('Terrorist Group')
plt.show()
```

```python
#here, we have defined the region names to region code from codebook provided in data sourcee
regions = {
    1: 'North America',
    2: 'Central America & Caribbean',
    3: 'South America',
    4: 'East Asia',
    5: 'Southeast Asia',
    6: 'South Asia',
    7: 'Central Asia',
    8: 'Western Europe',
    9: 'Eastern Europe',
    10: 'Middle East & North Africa',
    11: 'Sub-Saharan Africa',
    12: 'Australasia & Oceania'
}

df['Region Name'] = df['region'].map(regions)

#plot the number of attacks by region name
plt.figure(figsize=(10, 6))
region_attacks = df['Region Name'].value_counts()
sns.barplot(x=region_attacks.index, y=region_attacks.values)
plt.title('Number of Terrorist Attacks by Region')
plt.xlabel('Region')
plt.ylabel('Number of Attacks')
plt.xticks(rotation=90)
plt.show()
```

```python
# Plot the number of attacks per type of target
plt.figure(figsize=(12, 6))
target_attacks = df['Target Type'].value_counts()
sns.barplot(x=target_attacks.values, y=target_attacks.index, palette='viridis')
plt.title('Number of Terrorist Attacks by Type of Target')
plt.xlabel('Number of Attacks')
plt.ylabel('Type of Target')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
attacks_per_year = df['Year'].value_counts().sort_index()
plt.plot(attacks_per_year.index, attacks_per_year.values, marker='o', linestyle='-')
plt.title('Trend of Terrorist Attacks Over the Years')
plt.xlabel('Year')
plt.ylabel('Number of Attacks')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
year_casualties = df.groupby('Year')['Total_Casualties'].sum()
sns.lineplot(x=year_casualties.index, y=year_casualties.values, marker='o')
plt.title('Total Casualties in Terrorist Attacks Over the Years')
plt.xlabel('Year')
plt.ylabel('Total Casualties')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
month_attacks = df['Month'].value_counts().sort_index()
sns.lineplot(x=month_attacks.index, y=month_attacks.values, marker='o')
plt.title('Terrorist Attacks by Month')
plt.xlabel('Month')
plt.ylabel('Number of Attacks')
plt.xticks(range(1, 13))
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
year_casualties = df.groupby('Year')['Total_Casualties'].sum()
sns.barplot(x=year_casualties.index, y=year_casualties.values, palette='rocket')
plt.title('Total Casualties in Terrorist Attacks by Year')
plt.xlabel('Year')
plt.ylabel('Total Casualties')
plt.xticks(rotation=45)
plt.show()
```
